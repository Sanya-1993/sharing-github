```python
import numpy as np
```

```python
Income['NCHLT5'].describe()
```

```python
Income['Inc_M']=np.where(Income['SEX']==1,Income['INCWAGE'],0)
Income['Inc_F']=np.where(Income['SEX']==2,Income['INCWAGE'],0)
Income['Agelt30']=np.where(Income['AGE']<=30,1,0)
Income['Prop_F']=np.where(Income['SEX']==2,1,0)
Income['College']=np.where(Income['EDUC']>=7 ,1,0)
Income['Mar']=np.where(Income['MARST']<=2,1,0)
Income['Child5']=np.where(Income['NCHLT5']>0,1,0)
Income['Child']=np.where(Income['NCHILD']>0,1,0)



for i in range(1,6):
    Income[i]=np.where(Income['RACE']==i,1,0)
```

```python
#Income.loc[Income['Constr']==1]
```

```python
#Income['Constr'].unique()
```

```python
Income['Constr']=Income['OCC'].astype(str).str[:1]
```

```python

Income['Constr']=np.where(Income['Constr']=='6',1,0)
```

```python
Income_agg=Income.groupby(['YEAR','STATEFIP']).agg({'Inc_F':'mean','Inc_M':'mean','Agelt30':'sum','Child':'sum','Child5':'sum',1:'sum',2:'sum',3:'sum',4:'sum',5:'sum','SEX':'count','Prop_F':'sum','College':'sum','Mar':'sum','Constr':'sum'}).reset_index()
```
