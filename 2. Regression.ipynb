```python
#Reading party color data
Color=pd.read_excel("C:/Users/Sanya/Documents/Python/Python_Group/Data_variables.xlsx",sheet_name="x3")
```

```python
#Taking max votes for each year

Color['Max'] = Color.groupby(["year","state","totalvotes"],as_index=False)['candidatevotes'].transform('max')
Colorc=Color.loc[Color['Max']==Color['candidatevotes']]

Colorc=Colorc[['state_fips','year','party_simplified','state']]
Colorc
```

```python
#Defining party color as per latest election results eg: for 2009-2011,election results will be as per 2008 presidential election outcome
def ap(y,s,e):
    Color_2012=Colorc[Colorc['year']==y]
    Color_app=Color_2012
    for i in range(s,e):
        Color_new=Color_2012.drop(columns='year')
        Color_new['year']=i
        Color_app=Color_app.append(Color_new)
    return Color_app

    
```

```python
Color_12=ap(2012,2013,2016)
Color_9=ap(2008,2009,2012)
Color_16=ap(2016,2017,2020)
```

```python

#Appending all data
Color_all=pd.concat([Color_9,Color_12,Color_16],axis=0)
```

```python
#OLS Regression


def reg(op,form):
    ols=smf.ols(formula=form,data=GWG1).fit()
    path="Out/Reg/"+op+".csv"
    print(ols.summary())

    return ols 


#Trying different models

m1=reg("NoStateFE_NoInter_OnlyAge","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Empt+Prop_F+College+Constr+Mar+log_GDP")
m2=reg("NoStateFE_NoInter_OnlyChild5","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Child5+White+Black+AmIndian_AlaskaNative+Chinese+Japanese+Empt+Prop_F+College+Constr+Mar+log_GDP")
m3=reg("NoStateFE_NoInter_OnlyGDP","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Prop_F+College+Constr+Mar+log_GDP")
m4=reg("NoStateFE_NoInter_OnlyEmpt","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Prop_F+College+Constr+Mar+Empt")
m5=reg("NoStateFE_NoInter_All","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Prop_F+College+Constr+Mar+Empt+log_GDP")

m6=reg("NoStateFE_Inter","GWG_calc~C(YEAR)+DEMOCRAT*Pay_Transp+Agelt30+Child++White+Empt+Prop_F+College+Constr+Mar+log_GDP")
m7=reg("NoStateFE_NoInter","GWG_calc~C(YEAR)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Empt+Prop_F+College+Constr+Mar+log_GDP")
m8=reg("StateFE_Inter","GWG_calc~C(YEAR)+C(state)+DEMOCRAT*Pay_Transp+Agelt30+Child+White+Empt+Prop_F+College+Constr+Mar+log_GDP")
m9=reg("StateFE_NoInter","GWG_calc~C(YEAR)+C(state)+DEMOCRAT+Pay_Transp+Agelt30+Child+White+Empt+Prop_F+College+Constr+Mar+log_GDP")


s1=summary_col([m3,m4,m5],stars=True,float_format='%0.2f',
                  model_names=['OnlyGDP', 'OnlyEmpt', 'All'],
                  info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                             'R2':lambda x: "{:.2f}".format(x.rsquared)})



s1.tables[0].to_excel("Out/Reg/Multicoll_chk.xlsx")

s2=summary_col([m6,m7],stars=True,info_dict={'R-squared' : lambda x: "{:.2f}".format(x.rsquared),
           'No. observations' : lambda x: "{0:d}".format(int(x.nobs))})

s2.tables[0].to_excel("Out/Reg/NoStateFE.xlsx")



s3=summary_col([m8,m9],stars=True,info_dict={'R-squared' : lambda x: "{:.2f}".format(x.rsquared),
           'No. observations' : lambda x: "{0:d}".format(int(x.nobs))})
s3.tables[0].to_excel("Out/Reg/StateFE.xlsx")

```
